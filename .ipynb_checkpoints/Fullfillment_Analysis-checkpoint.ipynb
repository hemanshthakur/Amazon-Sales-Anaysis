# Initial Prep
## Importing Data and Libraries

```python
# Importing Essential Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
import matplotlib.dates as mdates
df = pd.read_csv('Cleaned_Amazon_Sales_Data.csv')
```

## Fulfilment Distribution by Merchants

```python
fulfillment_counts = df['Fulfilment'].value_counts().reset_index()
fulfillment_counts.columns = ['Fulfilment', 'Count']
plt.figure(figsize=(10, 6))
plt.pie( x='Count', data=fulfillment_counts, autopct='%1.1f%%',labels=fulfillment_counts['Fulfilment'])
plt.title('Fulfillment Methods Distribution')
plt.show()
```

### Insights
The pie chart shows that Amazon is the most popular fulfillment provider, with 69.6% of orders being shipped through Amazon. Merchant fulfillment accounts for the remaining 30.4% of orders.
### Recommendations:
Given that Amazon is the dominant fulfillment provider, it might be wise to prioritize building a strong relationship with them to ensure they continue to meet your fulfillment needs. You could also explore ways to leverage Amazon services to improve your fulfillment efficiency or reduce costs.

## Fulfilment Performance of Distributer by Delivery Status


```python
fulfilment_courier = df.groupby(['Fulfilment','Courier Status'])['Qty'].sum().reset_index()
fulfilment_courier.columns=['Fulfilment','Courier Status','Qty']
fulfilment_courier.value_counts()
sns.barplot( x='Fulfilment',y='Qty', hue='Courier Status', data=fulfilment_courier)
plt.title("Fulfilment Performance of Distributer by Delivery Status")
plt.show()
```

### Insights

Overall, Amazon delivers more packages than third-party merchants.<br>
The majority of Amazon's packages are shipped and "Unshipped".<br>
A smaller portion of Amazon's packages are "Cancelled" or "On the way".<br>
There are more "Unshipped" packages from Amazon than "Cancelled" packages.<br>
Third-party merchants have a similar distribution of delivery statuses, with most packages being "Shipped" and "Unshipped".<br>
There are also some "Cancelled" and "Unshipped" packages from third-party merchants.<br>

### Recommendations
**Prefer to use Amazon as fulfilment provider:** More people trust amazon with their after sales services due to reliability and safety, So it would be better to just full send whole fulfilment operations to Amazon, which will help out merchants as well because they do not have to cover for extra cost. 

## Fulfilment Performance of Distributer by Product Catagory

```python
fulfilment_wCategory = df.groupby(['Fulfilment','Category'])['Qty'].sum().reset_index()
fulfilment_wCategory.columns=['Fulfilment','Category','Qty']
fulfilment_wCategory.value_counts()
sns.barplot( x='Fulfilment',y='Qty', hue='Category', data=fulfilment_wCategory)
plt.legend(loc='upper right')
plt.title("Fulfilment Performance of Distributer by Category")
plt.show()
```

### Insights
Overall Looking at each category's sales ratio across both providers it seems to be 2.5 to 1 to Amazon, Amazon overall has better fulfilment results compared to merchants

### Recommendation
Improvement in customer Support is highly required by Merchant to be provided

##  Courier Distribution Status

```python
courier_status_counts = df['Courier Status'].value_counts().reset_index()
courier_status_counts.columns = ['Courier Status', 'Count']
plt.figure(figsize=(10, 6))
sns.barplot(x='Courier Status', y='Count', data=courier_status_counts)
plt.title('Courier Status by Distribution')
plt.xlabel('Courier Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
