# Initial Phase
## Importing Cleaned Data and Essential Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
import matplotlib.dates as mdates

df = pd.read_csv('Cleaned_Amazon_Sales_Data.csv')
```

# Product Analysis
## Distribution of product categories

```python
#Counting Total Items By Categories
category_counts = df['Category'].value_counts().reset_index()

#Naming The Columns 
category_counts.columns = ['Category', 'Count']

#Plotting Bar Chart
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x='Category', y='Count', data=category_counts)

# Adding title, labels, and grid
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')

max_count = category_counts['Count'].max() * 1.1
plt.ylim(bottom=0, top=max_count)

for row in category_counts.itertuples():
    category = row.Category  
    count = row.Count        
    ax.text(row.Index, count + count * 0.05, f"{count:.0f}", ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.show()
```

## Product Distribution by Sizes

```python
size_counts = df['Size'].value_counts().reset_index()
size_counts.columns = ['Size', 'Count']
plt.figure(figsize=(10, 6))
wedgeprops = dict(width=0.8) 
plt.pie(size_counts['Count'], startangle=140, wedgeprops=wedgeprops, pctdistance=0.8, autopct='%1.1f%%')
plt.legend(size_counts['Size'], loc='upper left', bbox_to_anchor=(1, 1),title='Sizes')
plt.title('Product Size Distribution')
plt.show()
```

## Quantity sold per category and size

```python
quantity_category_size=df.groupby(['Category', 'Size'])['Qty'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Qty', hue='Size', data=quantity_category_size)
plt.title('Quantity Sold per Category per Size')
plt.xlabel('Category')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```
